### Linear Regression using OLS (MLE)

$$W^{*} = (X^TX)^{-1}X^TY \;\;\;(Normal Equation)$$

- Assume:
    - $X$ and $Y$ are jointly normal distributed.
    - $\phi$ is the identity $\phi(X) = X$
    

- If the algorithm you want to use is Linear Model:
    - If number of features is small.
    - If number of training examples is small (i.e. around 20,000).

- Normal Equation is a good algorithm to consider to build your machine learning model.

```python
class OLS:
    def __init__(self):
        pass
    
    def fit(self, X, y):
        N = X.shape[0]
        # add cloumn of ones in X for the bias term
        X = np.append(X, np.ones((N, 1)), axis=1)        
        y = y.reshape(N, 1)
        # The Normal Equation (X^T X)^-1 X^T Y
        self.coef = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))

    def predict(self, X):
        X = np.append(X, np.ones((X.shape[0],1)), axis=1)
        preds = np.dot(X, self.coef)
        return preds
```

```python
ols_model = OLS()
ols_model.fit(X, Y)
```

```python
# get the estimated coefficient
theta = ols_model.coef
```

```python
# getting the predictions on X using the predict function
preds = ols_model.predict(X)
```

```python
def plot_regression_line(X, Y, ax, v=[30, 0]):
    
    plt.rcParams["figure.figsize"] = [7.00, 3.50]
    plt.rcParams["figure.autolayout"] = True

    x = np.linspace(-3, 3, 10)
    y = np.linspace(-3, 3, 10)

    x, y = np.meshgrid(x, y)
    eq = theta[0] * x + theta[1] * y + theta[-1]

    ax.plot_surface(x, y, eq, alpha=0.3)
    ax.scatter(X[:, 0], X[:, 1], Y, color='b')
    ax.view_init(v[0], v[1])
    ax.set_zlim(-5, 7)

    return ax
```

```python
fig = plt.figure(figsize=(20, 8))
plt.suptitle('3-D view of the data points an the regression line from different POV', fontsize=18)
ax1 = fig.add_subplot(1, 3, 1, projection='3d')
ax2 = fig.add_subplot(1, 3, 2, projection='3d')
ax3 = fig.add_subplot(1, 3, 3, projection='3d')
plot_regression_line(X, Y, ax1, [50, 0])
plot_regression_line(X, Y, ax2, [20, 10])
plot_regression_line(X, Y, ax3, [40, 50]);
```
